## Quality-Diversity tradeoff

```python
def build_df(ref_path, gen_path, human_path, decoders):
    #load referencens and generations
    refs = pkl.load(open(ref_path, 'rb'))
    gen_probs = pkl.load(open(gen_path, 'rb'))
    decoder_to_id = {dec : i for i, dec in enumerate(decoders)}

    df = pd.read_csv(human_path)
    human_ids =  np.sort(df.example_id.unique())
    mean_ratings = df.groupby(['example_id', 'decoder'], sort=True)['median'].mean().reset_index(drop=False)
    mean_ratings.rename(columns={'median' : 'human score'}, inplace=True)
    #construct generations data frame
    example_id = np.repeat(list(range(400)), len(decoders))
    dec = decoders * 400
    gen_nll = [x[1.0].sum().item() * (-np.log2(np.exp(1))) for x in gen_probs['log_probs']]
    gen_len = [x[1.0].shape[-1] for x in gen_probs['log_probs']]
    gens_df = pd.DataFrame({'example_id' : example_id, 'decoder' : dec, 'nll' : gen_nll, 'len' : gen_len})

    #construct references data frame
    dec = 'reference'
    example_id = list(range(400))
    ref_nll = [x[1.0].sum().item() * (-np.log2(np.exp(1))) for x in refs['log_probs']]
    ref_len = [x[1.0].shape[-1] for x in refs['log_probs']]
    refs_df = pd.DataFrame({'example_id' : example_id, 'decoder' : dec, 'nll' : ref_nll, 'len' : ref_len})

    combined = pd.concat((gens_df, refs_df), axis=0)
    data = mean_ratings.merge(combined, on=['example_id', 'decoder'], how='left')
    
    return data
```

## $\delta$ - estimates

```python
def get_topn(df, n_best, ignore_reference=True):
    if ignore_reference: 
        top_n = df[df.decoder != 'reference'].groupby('example_id')['human score'].nlargest(n_best).reset_index()
    else:
        top_n = df.groupby('example_id')['human score'].nlargest(n_best).reset_index() 
    top_n.rename(columns={'level_1' : 'old_index'}, inplace=True)
    top_n = df.iloc[top_n['old_index'], :].copy()
    
    return top_n

def get_bottomn(df, n_best, ignore_reference=True):
    if ignore_reference: 
        top_n = df[df.decoder != 'reference'].groupby('example_id')['human score'].nsmallest(n_best).reset_index()
    else:
        top_n = df.groupby('example_id')['human score'].nsmallest(n_best).reset_index() 
    top_n.rename(columns={'level_1' : 'old_index'}, inplace=True)
    top_n = df.iloc[top_n['old_index'], :].copy()
    
    return top_n

def plot_topn(cum_batches, sample_means, df, top_n, w=2, h=8, figsize=(15, 3), *args, **kwargs):
    fig, axs = plt.subplots(w, h, sharex=True, sharey=True, figsize=figsize)
    ranks = ['top ' + str(i + 1) for i in range(n_best)]
    #hack for ULG
    if not isinstance(axs, np.ndarray):
        axs = np.array([axs])
    for k, ax in tqdm(enumerate(axs.flatten()), total=w*h):
        current_batch = cum_batches[k, :]
        data = pd.DataFrame({'nll' : current_batch[~np.isclose(current_batch, 0)]})
        g = sns.kdeplot(data=data, x='nll', fill=True, cut=0, ax=ax, common_norm=False, *args,
                       **kwargs)
        g.axvline(sample_means[k], color=pal[0], label='mean', linestyle='--', alpha=1)
        g.axvline(df[df.decoder=='reference'].nll.iloc[k], color=pal[1], label='ref', linestyle='--', alpha=1)
        for rank, nll in enumerate(top_n.nll[k * n_best : (k + 1) * n_best]):
            g.axvline(nll, color=pal[rank + 2], label=ranks[rank], linestyle='--', alpha=1)

    #legend
    #remove duplicates
    handles, labels = plt.gca().get_legend_handles_labels()
    fig.legend(handles[:n_best + 2], labels[:n_best + 2])
    
def get_delta(top_n, cum_batches):
    #delta
    max_diff = top_n.groupby('example_id')['diff'].max()
    mean_diff = top_n.groupby('example_id')['diff'].mean()
    median_diff = top_n.groupby('example_id')['diff'].median()
    
    #sample_means = cum_batches.mean(axis=-1)
    sample_std = cum_batches.std(axis=-1)

    delta_frame = pd.DataFrame({'max' : max_diff, 'mean' : mean_diff, 'median' : median_diff, 'std' : sample_std})
    prob_mass = pd.DataFrame()

    for method in delta_frame.columns:
        tmp = []
        for i, sample in enumerate(cum_batches):
            left = sample.mean() - delta_frame[method].iloc[i]
            right = sample.mean() + delta_frame[method].iloc[i]
            n = sample.shape[-1]
            m = sample[(sample <= right) & (sample >= left)].shape[-1]
            tmp.append(100 * m / n)
        prob_mass[method] = tmp
    
    return delta_frame, prob_mass
```

### Story Generation

```python
n_samples = 100
n_examples = 200
ignore_reference = True
n_best = 3

task = 'stories'
ref_path = f'../data/forward/{task}/{task}_refs_probs.p'
gen_path = f'../data/forward/{task}/{task}_gens_probs.p'
human_path = '../data/human_scores/stories_medium.final.csv'
samples_path = '../data/samples/stories_medium.test.bayes.generations.p'

decoders = ['greedy', 'ancestral', 'beam', 'nucleus', 'diverse_beam', 'top_k', 'bayes', 'beam_large']

df = build_df(ref_path, gen_path, human_path, decoders)
data = pkl.load(open(samples_path, 'rb'))['bayes']
cum_batches = [data['cum'][i: i + n_samples] for i in range(0, n_examples * n_samples, n_samples)]
cum_batches = np.array(cum_batches) * (-np.log2(np.exp(1)))
sample_means = cum_batches.mean(axis=-1)
sample_std = cum_batches.std(axis=-1)

top_n = get_topn(df, n_best, ignore_reference)
top_n = top_n.iloc[:(n_best * n_examples), :]
top_n['sample_mean'] = np.repeat(sample_means, n_best)
top_n['sample_std'] = np.repeat(sample_std, n_best)
top_n['diff'] = (top_n['sample_mean'] - top_n['nll']).abs()

delta, prob = get_delta(top_n, cum_batches)
```

```python
#normalized by length
def fun(lst):
    nump = np.array(lst)
    t = np.isclose(nump, 0.0)
    return nump[~t].shape[0]

frame = df.copy()
display = ['reference', 'top 1', 'bottom 1']

sample_lengths = list(map(fun, data['log_prob']))
len_batches = np.array([sample_lengths[i: i + n_samples] for i in range(0, n_examples * n_samples, n_samples)])
sample_len_corrected = np.divide(cum_batches,
                                 len_batches, out=np.zeros_like(cum_batches), where=len_batches!=0)
sample_avg = sample_len_corrected.mean(axis=-1)

top_n = get_topn(frame, n_best, ignore_reference)
top_n.loc[:, 'decoder'] = 'top 1'

worst_n = get_bottomn(frame, n_best, ignore_reference)
worst_n.loc[:, 'decoder'] = 'bottom 1'

frame = pd.concat((frame, top_n, worst_n)).reset_index(drop=True)
frame['avg'] = frame.nll / frame.len

frame = frame[frame.decoder.isin(display)]

figsize = (5, 3.0)
log_scale = True
fig, axs = plt.subplots(1, 1, figsize=figsize)
order=list(frame.decoder.unique())[:5]

g = sns.kdeplot(data=frame, x='avg', hue='decoder', hue_order=order[::-1], fill=True, common_norm=False,
   alpha=.5, linewidth=3, log_scale=log_scale, ax=axs, legend=False)
avg_flat = sample_len_corrected.flatten()
flat = cum_batches.flatten()
samp = pd.DataFrame({'nll' : flat[~np.isclose(flat, 0.0)], 'decoder' : 'model',
                     'avg' : avg_flat[~np.isclose(avg_flat, 0.0)]})
f = sns.kdeplot(data=samp, x='avg', fill=True, common_norm=False,  
                alpha=.5, linewidth=3, log_scale=log_scale, ax=axs, legend=True, color=pal[len(order)+1])


#axs.legend()
plt.legend(labels=order + ['model'], loc='upper left', frameon=False, fontsize=14)

plt.axvline(samp.avg.mean(), linestyle='-.', linewidth=1, color='black')

plt.xlabel(r'Average Information $\frac{I(y)}{|y|}$', fontsize=17)
plt.ylabel(r'Density', fontsize=17)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
# Hide the right and top spines
axs.spines['right'].set_visible(False)
axs.spines['top'].set_visible(False)
#plt.savefig('plots2/AS_avg_info.jpeg', dpi=300, bbox_inches='tight')
#samples -> distribution
#figure 2
```

```python
def build_special_df(ref_path, gen_path, human_path, decoders, threshold=2):
    const = (-np.log2(np.exp(1)))
    #load referencens and generations
    refs = pkl.load(open(ref_path, 'rb'))
    gen_probs = pkl.load(open(gen_path, 'rb'))
    decoder_to_id = {dec : i for i, dec in enumerate(decoders)}

    df = pd.read_csv(human_path)
    human_ids =  np.sort(df.example_id.unique())
    mean_ratings = df.groupby(['example_id', 'decoder'], sort=True)['median'].mean().reset_index(drop=False)
    mean_ratings.rename(columns={'median' : 'human score'}, inplace=True)
    #construct generations data frame
    example_id = np.repeat(list(range(400)), len(decoders))
    dec = decoders * 400
    gen_nll = [x[1.0][x[1.0].isfinite()].sum().item() * const if \
               ((x[1.0].shape[-1] - x[1.0].isfinite().sum().item()) <= threshold) else np.nan \
              for x in gen_probs['log_probs']]
    
    gen_len = [x[1.0].isfinite().sum().item() if \
               ((x[1.0].shape[-1] - x[1.0].isfinite().sum().item()) <= threshold) else np.nan \
              for x in gen_probs['log_probs']]
    gens_df = pd.DataFrame({'example_id' : example_id, 'decoder' : dec, 'nll' : gen_nll, 'len' : gen_len})

    #construct references data frame
    dec = 'reference'
    example_id = list(range(400))
    ref_nll = [x[1.0][x[1.0].isfinite()].sum().item() * const if \
               ((x[1.0].shape[-1] - x[1.0].isfinite().sum().item()) <= threshold) else np.nan \
              for x in refs['log_probs']]
    ref_len = [x[1.0].isfinite().sum().item() if \
               ((x[1.0].shape[-1] - x[1.0].isfinite().sum().item()) <= threshold) else np.nan \
              for x in refs['log_probs']]
    refs_df = pd.DataFrame({'example_id' : example_id, 'decoder' : dec, 'nll' : ref_nll, 'len' : ref_len})

    combined = pd.concat((gens_df, refs_df), axis=0)
    data = mean_ratings.merge(combined, on=['example_id', 'decoder'], how='left')
    
    return data

def get_special_topn(df, n_best, ignore_reference=True):
    if ignore_reference: 
        top_n = df[df.decoder != 'reference'].groupby(['example_id', 'transform'])['human score']\
        .nlargest(n_best).reset_index()
    else:
        top_n = df.groupby('example_id')['human score'].nlargest(n_best).reset_index() 
    top_n.rename(columns={'level_2' : 'old_index'}, inplace=True)
    top_n = df.iloc[top_n['old_index'], :].copy()
    
    return top_n

def get_special_bottomn(df, n_best, ignore_reference=True):
    if ignore_reference: 
        top_n = df[df.decoder != 'reference'].groupby(['example_id', 'transform'])['human score']\
        .nsmallest(n_best).reset_index()
    else:
        top_n = df.groupby('example_id')['human score'].nsmallest(n_best).reset_index() 
    top_n.rename(columns={'level_1' : 'old_index'}, inplace=True)
    top_n = df.iloc[top_n['old_index'], :].copy()
    
    return top_n
```
